```python
import numpy as np
from gensim.models import Word2Vec
#from sklearn.manifold import TSNE
#import matplotlib.pyplot as plt
```

```python
text = open('newspaper.txt', 'r', encoding='utf8')  # türkçe karakterler için encoding = "utf8" i kullanmamız önemli
text = text.read()
list_sent = text.split('\n')

corpus = []

for sent in list_sent:
    corpus.append(sent.split())
```

```python
print(corpus[:10])

```

```python
text[:1000]
```

```python
list_sent[:10]
```

```python
model = Word2Vec(corpus, vector_size=100, window=5, min_count=5, sg=1)

# vector_size=100 : vektörümüzün kaç boyutlu olmasını istiyorsak. (bazı sürümlerde "size" olarak da geçiyor)

# min_count=5: bir token 5 ten fazla ise eğitime dahil et. az ise eğitime dahil etme

# skip-gram küçük corpuslarda iyi sonuç verirken, CBOW çok daha büyük corpuslarda iyi sonuç veriyor.
```

```python
model.wv['ankara']

# ankara için bana 100 boyutlu bir vektör verdi.
```

```python
model.wv.most_similar('birşey')

# eğittiğim corpus a göre, "birşey" kelimesi ile anlamsal yakınlığa sahip olan kelimeleri buluyor. 
  # yanındaki değerler de ne kadar ilişkili olduğunu gösteriyor.. 

# Similar; anlamsal ilişkiyi bulurken yan yana geçen kelimeleri  ve yerine ikame edilen kelimeleri ayrı ayrı değerlendiriyor.
    # bunu yaparken önce yerine ikame edilen kelimelere bakıyor. onu similar olarak kaydediyor. 
        # eğer ikame kelime bulamazsa yan yana geçen kelimelere bakarak benzerlik ağırlığını belirliyor
```

```python
model.wv.most_similar('eve')

# "eve" ile birlikte kullanılan kelimeler ile hangi başka kelimeler birlikte kullanılıyor ise onlar eve nin yerine ikame edilen kelimelerdir.
  # mesela eve gittim i ele alıp, gittim le birlikte kullanılan diğer kelimeleri aşağıda sıraladığını görüyoruz. 
   # (dükkana gittim, ormana gittim, mağazaya gittim gibi)    
```

```python
model.wv.most_similar('mavi')
```

```python
model.wv.most_similar(positive=['öğrenme', 'doktor'], negative=['tedavi'], topn=1)
```

```python
model.wv.most_similar(positive=['ankara', 'rusya'], negative=['moskova'], topn=1)
```

```python
model.save("word2vec.model")
```

```python
from gensim.scripts.glove2word2vec import glove2word2vec
from gensim.models import KeyedVectors
```

```python
glove_model = 'glove.6B.100d.txt'  # bu eğitilmiş hazır model. bunu kullanıyoruz.
word2vec = 'glove.6B.100d.glove'
glove2word2vec(glove_model, word2vec)


# glove modeli word2vec e dönüştürüyoruz.
# glove2word2vec(glove_model, word2vec) --> glove modeli al word2vec e çevir. demek
```
